```python
import requests
from bs4 import BeautifulSoup

# Step 1: Grab the web page
url = "http://quotes.toscrape.com/"
response = requests.get(url)

# Step 2: Check if it worked
print("Status Code:", response.status_code)

# Step 3: Print a tiny bit of the HTML
print(response.text[:500])  # preview the first 500 characters
```

```python

soup = BeautifulSoup(response.text, 'html.parser')


quote_elements = soup.find_all('div', class_='quote')


for quote in quote_elements:
    text = quote.find('span', class_='text').get_text()
    author = quote.find('small', class_='author').get_text()
    print(f"{text} — {author}")
```

```python
# Grab book page
book_url = "http://books.toscrape.com/"
book_response = requests.get(book_url)

# Check response
print("Status Code:", book_response.status_code)
print(book_response.text[:500])
```

```python
# Parse the page using BeautifulSoup
book_soup = BeautifulSoup(book_response.text, 'html.parser')

# Find all book containers
books = book_soup.find_all('article', class_='product_pod')

# Loop through each book and print title + price
for book in books:
    title = book.h3.a['title']
    price = book.find('p', class_='price_color').text
    print(f"{title} — {price}")
```

```python
import pandas as pd

# Create lists to hold data
titles = []
prices = []

for book in books:
    titles.append(book.h3.a['title'])
    prices.append(book.find('p', class_='price_color').text)

# Create a DataFrame
df = pd.DataFrame({
    'Title': titles,
    'Price': prices
})

# Save to CSV
df.to_csv('books.csv', index=False)

print("Saved books.csv with", len(df), "books!")
```
